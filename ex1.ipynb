```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
means = [2, 2]
cov = [[1, 0], [0, 1]]
N = 50
X = np.random.multivariate_normal(means, cov, N)
```

```python
fig, axes = plt.subplots()
axes.scatter(X[:, 0], X[:, 1])
```

```python
X = X - X.mean(axis=0)
X_std = X/ np.std(X, axis=0)
cov_mat = np.cov(X_std.T)
```

```python
e, v = np.linalg.eig(cov_mat)
print('E-value:', e)
print('E-vector', v)
```

```python
max_vec = v[:, e.argmax()]
print(max_vec)
```

```python
dt = X_std.dot(max_vec)*max_vec[:, np.newaxis]
```

```python
plt.scatter(dt[0], dt[1])
```

```python
from sklearn.preprocessing import StandardScaler
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
df = pd.read_csv(url, names=['sepal length','sepal width','petal length','petal width','target'])
```

```python
features = ['sepal length', 'sepal width', 'petal length', 'petal width']
x = df.loc[:, features].values
```

```python
fig, axes = plt.subplots()
axes.scatter(x[:, 0], x[:, 1])
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(x)
```

```python
plt.scatter(principalComponents[:,0], principalComponents[:,1])
```

```python
plt.plot(principalComponents[:,0], principalComponents[:,1], 'bo', markersize = 4, alpha = .8)
plt.axvline(x=0, ymin=0, ymax=1, lw=1.5, label='boudary', c='gray')
plt.axhline(y=0, xmin=0, xmax=1, lw=1.5, c='gray')
plt.show()
```
